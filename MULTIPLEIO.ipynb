```python
# CS440_MP01_V1
# Team Member:Zhenwei Wu, Yuji Chai, Zhoushi Zhu
# function list
# drawMaze(ls): print out maze;
# heuristic(ls,end): generate distance from target(or 'food'), return a 2D list
# drawSol(ls,map): draw solution on maze(or 'origin_mazeList')
# bfs(ls, start), dfs(ls, start), A_star(ls),
# 1) open text file by fp; 
# 2) copy text file into mazeList[], use in path searching; 
# 3) deepcopy origin_mazeList[] from mazeList[], use in drawSol(ls,map).
fp = './bigDots.txt'
mazeList=[]
origin_mazeList=[]
mazeMapList=[]
row=0
column=0
ct=0
pecman=0
foodList = []
TOTAL = 0
foodMap = {}

with open(fp) as f:  
    for line in f:
        row+=1
        column=len(line)
        temp=[]
        for a in line:
            if(a=='%'):
                temp.append('%')
                ct+=1
            elif(a=='P'):
                temp.append('P')
                ct+=1
                pecman=ct
            elif(a=='.'):
                temp.append('.')
                ct+=1
                TOTAL+=1
            elif(a==' '):
                temp.append(' ')
                ct+=1
        mazeList.append(temp)
        
import copy
origin_mazeList=copy.deepcopy(mazeList)


def drawMaze(ls):
    for i in ls:
        temp=""
        for j in i:
            temp+=j
        print(temp)
        
        
drawMaze(mazeList)
print("row: ",row," ;column: ",column)
print("pecman: ",pecman," ;food: ",TOTAL)
```

```python
from copy import deepcopy

def Uniform_bfs(ls):
    # 'start' is the index; int(start/column) is the row number; start%column-1 is the column number
    # ls[int(start/column)][start%column-1] = 'P'
    stack=[pecman]
    backup = deepcopy(ls)
    mazeMap={}
    count = 0
    foodList.append(pecman)
    while stack and count < TOTAL:
        ind=stack.pop(0)
        #calculate row as 'r', column as 'c', find current content use ls[r][c]
        r=int(ind/column)
        c=ind%column-1
        cur=ls[r][c]
        if cur != '%' and cur != 'V':
            if cur is '.':
                print(ind)
                #print(len(backup))
                backup[r][c] = ' '
                mazeMapList.append(deepcopy(mazeMap))
                foodList.append(ind)
                mazeMap.clear()
                ls = deepcopy(backup)
                stack = [ind]
                count+=1
                continue

            else:
                ls[r][c]='V'
                if (r+1)<row:
                    if ls[r+1][c] != 'V' and ls[r+1][c] != '%':
                        temp=ind+column
                        mazeMap[temp]=ind
                        #print(r+1,',',c,ls[r+1][c])
                        stack.append(temp)
                # up
                if (r-1)>=0:
                    if ls[r-1][c] != 'V' and ls[r-1][c] != '%':
                        temp=ind-column
                        mazeMap[temp]=ind
                        #print(r-1,',',c,ls[r-1][c])
                        stack.append(temp)
                # right
                if (c+1)<column:
                    if ls[r][c+1] != 'V' and ls[r][c+1] != '%':
                        temp=ind+1                  
                        mazeMap[temp]=ind
                        #print(r,',',c+1,ls[r][c+1])
                        stack.append(temp)
                # left
                if (c-1)>=0:
                    if ls[r][c-1] != 'V' and ls[r][c-1] != '%':
                        temp=ind-1                    
                        mazeMap[temp]=ind
                        #print(r,',',c-1,ls[r][c-1])
                        stack.append(temp)
                        
    print(count)
```

```python
Uniform_bfs(mazeList)
```

```python
# draw solution from food->pecman on mazeList; 
# need drawMaze(mazeList) to visualize path 
# ct: total steps
       
def drawSol(ls, start, end, solutionMap):
    x=end
    ct=0
    while x != start:
        #print(x)
        r=int(x/column)
        c=x%column-1
        ls[r][c]='*'
        x=solutionMap[x]
        ct+=1
    return ct

def drawSols(ls, foodList, solutionList):
    temp = copy.deepcopy(ls)
    count = 0
    temp_count = 0
    for i in range(len(solutionList)):
        
        temp_count = drawSol(temp, foodList[i], foodList[i + 1], solutionList[i])
        count += temp_count
        print("The solution for the ", i+1, "round:")
        print("This round costs ", temp_count, "steps.")
        drawMaze(temp)
        print("")
    print("The entire solution costs " , count , "steps.")
    return count
```

```python
drawSols(origin_mazeList ,foodList, mazeMapList)
```
