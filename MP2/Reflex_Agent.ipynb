```python
import numpy as np
board = np.zeros([7,7])
# board[0][14] = 1

def print_board(board):
    print('  0   1   2   3   4   5   6 ')
    print(' --- --- --- --- --- --- ---')
    for i in range(7):
        temp = []
        temp.append('|')
        for j in range(7):
            if board[i][j] == 0:
                temp.append(' ')
            elif board[i][j] == 1:
                temp.append('@')
            else:
                temp.append('#')
            temp.append('|')
        temp.append(str(i))
        print(' '.join(temp))
        print(' --- --- --- --- --- --- ---')
#         print(' -----------------------------------------------------------')
        
print_board(board)
                
```
