```python
def evaluation_heuristic(b, max_agent):
    v_map_horiz_1 = np.zeros([7,7])
    v_map_vertical_1 = np.zeros([7,7])
    v_map_rd_1 = np.zeros([7,7])
    v_map_ru_1 = np.zeros([7,7])
    
    v_map_horiz_2 = np.zeros([7,7])
    v_map_vertical_2 = np.zeros([7,7])
    v_map_ru_2 = np.zeros([7,7])
    v_map_rd_2 = np.zeros([7,7])
    
    for i in range(7):
        for j in range(7):
            v_map_horiz_1[i][j] = 1 if b[i][j] == max_agent else 0
            v_map_vertical_1[i][j] = 1 if b[i][j] == max_agent else 0
            v_map_rd_1[i][j] = 1 if b[i][j] == max_agent else 0
            v_map_ru_1[i][j] = 1 if b[i][j] == max_agent else 0
                
    
    find_winning_block_horiz(b, max_agent, v_map_horiz_1)
    find_winning_block_vertical(b, max_agent, v_map_vertical_1)
    find_winning_block_ru(b, max_agent, v_map_ru_1)
    find_winning_block_rd(b, max_agent, v_map_rd_1)
    
    for i in range(7):
        for j in range(7):
            v_map_horiz_1[i][j] = v_map_horiz_1[i][j]*v_map_vertical_1[i][j]*v_map_ru_1[i][j]*v_map_rd_1[i][j]
            
    n = max_agent%2 + 1
    for i in range(7):
        for j in range(7):
            v_map_horiz_2[i][j] = 1 if b[i][j] == n else 0
            v_map_vertical_2[i][j] = 1 if b[i][j] == n else 0
            v_map_rd_2[i][j] = 1 if b[i][j] == n else 0
            v_map_ru_2[i][j] = 1 if b[i][j] == n else 0
            
    
    find_winning_block_horiz(b, n, v_map_horiz_2)
    find_winning_block_vertical(b, n, v_map_vertical_2)
    find_winning_block_ru(b, n, v_map_ru_2)
    find_winning_block_rd(b, n, v_map_rd_2)
    
    for i in range(7):
        for j in range(7):
            v_map_horiz_2[i][j] = v_map_horiz_2[i][j]*v_map_vertical_2[i][j]*v_map_ru_2[i][j]*v_map_rd_2[i][j]
#     print('min',np.sum(v_map_horiz_2))
#     print('max',np.sum(v_map_horiz_1))
            
    for i in range(7):
        for j in range(7):
            v_map_horiz_1[i][j] = v_map_horiz_1[i][j] -v_map_horiz_2[i][j]
            
    return np.sum(v_map_horiz_1)
            
    
    
    
    
```

```python
import math
def Minimax(board, max_val):
    # this function would return the index = i * 15 + j
    
    # determine what is the max_val and min_val
    # max is the agent, min is the opponent
    # CASE1: max_val = 1, min_val = 2
    # CASE2: max_val = 2, min_val = 1
    if max_val == 1:
        min_val = 2
    elif max_val == 2:
        min_val = 1
    else:
        print('Minmax::max_val input is not correct!')
        return -1
    
    # First is max move, Second is the min move, Thrid is the max move
    expand_count = 0
    
    # Searching through all the first move
    move1_idx = -1
    move1_eval = -math.inf
    for idx1 in range(49):
        if board[idx1//7][idx1%7] == 0:
            expand_count += 1
            print(expand_count)
            # set the current move and find its evaluation
            board[idx1//7][idx1%7] = max_val
            
            # Searching through all the second move
            move2_idx = -1
            move2_eval = math.inf
            for idx2 in range(49):
                if board[idx2//7][idx2%7] == 0:
                    expand_count += 1
#                     print(expand_count)
                    # set the current move and find its evaluation
                    board[idx2//7][idx2%7] = min_val
    
                    #Searching through all the thrid move
                    move3_idx = -1
                    move3_eval = -math.inf
                    for idx3 in range(49):
                        if board[idx3//7][idx3%7] == 0:
                            expand_count += 1
                            
                            # set the current move and find its evaluation
                            board[idx3//7][idx3%7] = max_val
                            temp = evaluation_heuristic(board, max_val)
                            #temp = 0
                            
                            # if the current eval is larger
                            if temp > move3_eval:
                                move3_eval = temp
                                move3_idx = idx3
                            
                            # reset the current move
                            board[idx3//7][idx3%7] = 0
                    
                    # if the current eval is smaller
                    if move3_eval < move2_eval:
                        move2_eval = move3_eval
                        move2_idx = idx2
                            
                    # reset the current move
                    board[idx2//7][idx2%7] = 0
            
            # if the current eval is larger
            if move2_eval > move1_eval:
                move1_eval = move2_eval
                move1_idx = idx1
                            
            # reset the current move
            board[idx1//7][idx1%7] = 0
    
    print(expand_count)
    print(move1_eval)
    print('next = ',move1_idx)
    return move1_idx
```
