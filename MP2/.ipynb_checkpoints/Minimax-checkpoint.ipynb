```python
def check_five(board, target):
    r = target//7
    c = target%7
    if board[r][c] == 0:
        return -1 
    
    #check horizontal
    i = c
    count = 0

    while i >= 0:#left
        if board[r][i] == board[r][c]:
            count+=1
            i-=1
#             print("left")
        else:
            break 
            
    i = c
    count-=1
    
    while i < 7:#right
        if board[r][i] == board[r][c]:
            count+=1
            i+=1
#             print("right")
        else:
            break
    if count >= 5:
        return 1
        
    #check vertical
    i = r
    count = 0

    while i >= 0:#up
        if board[i][c] == board[r][c]:
            count+=1
            i-=1
#             print("up")
        else:
            break
        
    i = r
    count-=1
    while i < 7:#down
        if board[i][c] == board[r][c]:
            count+=1
            i+=1
#             print("down")
        else:
            break

    if count >= 5:
        return 1
        
    #check left up and right down
    i = r 
    j = c
    count = 0
    
    while i >= 0 and j>=0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j-=1
#             print("left up")
        else:
            break

    i = r
    j = c
    count-=1
    while i < 7 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j+=1
#             print("down right")
        else:
            break
    if count >= 5:
        return 1
        
    #check left down and right up
    i = r 
    j = c
    count = 0

    while i < 7 and j >= 0:#left up
        if board[i][j] == board[r][c]:
            count+=1
            i+=1
            j-=1
#             print("left down")
        else:
            break

    i = r
    j = c
    count-=1
    while i >= 0 and j < 7:#right down
        if board[i][j] == board[r][c]:
            count+=1
            i-=1
            j+=1
#             print("right up")
        else:
            break
    if count >= 5:
        return 1
        
    return 0
```
