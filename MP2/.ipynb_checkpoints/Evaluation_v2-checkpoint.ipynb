```python
import math
def AlphaBeta(board, max_val):
    # this function would return the index = i * 15 + j
    
    # determine what is the max_val and min_val
    # max is the agent, min is the opponent
    # CASE1: max_val = 1, min_val = 2
    # CASE2: max_val = 2, min_val = 1
    if max_val == 1:
        min_val = 2
    elif max_val == 2:
        min_val = 1
    else:
        print('AlphaBeta::max_val input is not correct!')
        return -1
    
    # First is max move, Second is the min move, Thrid is the max move
    expand_count = 0
    
    # Searching through all the first move
    move1_idx = -1
    move1_eval = [-np.inf,-np.inf]
    for idx1 in range(49):
        if board[idx1//7][idx1%7] == 0:
            expand_count += 1
#             print(expand_count)
            # set the current move and find its evaluation
            board[idx1//7][idx1%7] = max_val
            
            # Searching through all the second move
            move2_idx = -1
            move2_eval = [np.inf, np.inf]
            for idx2 in range(49):
                if not less_than(move1_eval , move2_eval):
                    break
                if board[idx2//7][idx2%7] == 0:
                    expand_count += 1
                    
                    # set the current move and find its evaluation
                    board[idx2//7][idx2%7] = min_val
    
                    #Searching through all the thrid move
                    move3_idx = -1
                    move3_eval = [-np.inf,-np.inf]
                    for idx3 in range(49):
                        if less_than(move2_eval, move3_eval):
                            break
                        if board[idx3//7][idx3%7] == 0:
                            expand_count += 1
                            
                            # set the current move and find its evaluation
                            board[idx3//7][idx3%7] = max_val
                            temp = evaluation_wrapper(board, max_val)
                            #temp = 0
                            
                            # if the current eval is larger
                            if not less_than(temp , move3_eval):
                                move3_eval = temp
                                move3_idx = idx3
                            
                            # reset the current move
                            board[idx3//7][idx3%7] = 0
                    
                    # if the current eval is smaller
                    if less_than(move3_eval , move2_eval):
                        move2_eval = move3_eval
                        move2_idx = idx2
                            
                    # reset the current move
                    board[idx2//7][idx2%7] = 0
            
            # if the current eval is larger
            if not less_than(move2_eval , move1_eval):
                move1_eval = move2_eval
                move1_idx = idx1
                            
            # reset the current move
            board[idx1//7][idx1%7] = 0
#     print(expand_count)
#     print(move1_eval)
    print(move1_idx)
#     print(move2_eval)
#     print(move2_idx)
#     print(move3_eval)
#     print(move3_idx)
    return move1_idx
```

```python
def reflex_agent_wrapper(board, agent):
    next_step = [-1, 7]
#     print('***************************step 1*******************************')
    for j in range(49):
        r = j//7
        c = j%7
        if board[r][c] == agent:
            temp = check_four(board, r*7+c)
            if temp[0] == 1:
                if next_step[1] > temp[2]:
                    next_step[0] = temp[1]
                    next_step[1] = temp[2]
                    print(next_step)
                elif next_step[1] == temp[2]:
                    if next_step[0] < temp[1]:
                        next_step[0] = temp[1]
                        next_step[1] = temp[2]
                        print(next_step)
#     print('***************************step 2*******************************')
    if next_step[0] == -1:
        for j in range(49):
            r = j//7
            c = j%7
            if board[r][c] != agent and board[r][c] != 0:
                temp = check_four(board, r*7+c)
                if temp[0] == 1:
                    if next_step[1] > temp[2]:
                        next_step[0] = temp[1]
                        next_step[1] = temp[2]
                        print(next_step)
                    elif next_step[1] == temp[2]:
                        if next_step[0] < temp[1]:
                            next_step[0] = temp[1]
                            next_step[1] = temp[2]
                            print(next_step)
#     print('***************************step 3*******************************')
    if next_step[0] == -1:
        for j in range(49):
            r = j//7
            c = j%7
            if board[r][c] != agent and board[r][c] != 0:
                temp = check_three(board, r*7+c)
                if temp[0] == 1:

                    if next_step[1] > temp[2]:
                        next_step[0] = temp[1]
                        next_step[1] = temp[2]
                        print(next_step)
                    elif next_step[1] == temp[2]:
                        if next_step[0] < temp[1]:
                            next_step[0] = temp[1]
                            next_step[1] = temp[2]  
                            print(next_step)
#     print('***************************step 4*******************************')
    if next_step[0] == -1:
        r,c = find_winning_block(board, agent)
        if r != -1 and c != -1:
            next_step[0] = r
            next_step[1] = c
            print(next_step)
#     print('***************************step 5*******************************')  
    if next_step[0] == -1:
        for c in range(7):
            for r in range(7):
                if board[r][c] == 0:
                    next_step[0] = r
                    next_step[1] = c
                    print(next_step)
                    break
            if next_step[0] != -1:
                break
#     print('***************************step 6*******************************')
    board[next_step[0]][next_step[1]] = agent
    print(agent, 'next_step is', next_step)
    return next_step[0]*7+next_step[1]
```
